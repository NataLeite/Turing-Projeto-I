# Limpeza de dados
Antes de começarmos a análise de dados, devemos, primeiramente, verificar a existência de dados faltantes ou equivocados para "limpá-los" de nossos datasets. Iniciaremos, portanto, por essa etapa. Para facilitar este processo, uniremos os datasets pelo seguinte comando:

```python
uniao = pd.merge(treino, teste, how = 'outer')    # Une os dois dataframes em 'uniao'
uniao.info()
```

Considerando o número de linhas presentes nos conjuntos de dados de treino e de teste (54808 e 23490, respectivamente), é possível notar a existência de dados faltantes ('null') nas features 'education', 'no_of_trainings', 'previous_year_rating' e 'avg_training_score', já que, nestes itens, o número de dados verificáveis difere do total 78298. Portanto, teremos que analisar caso a caso para decidir o melhor procedimento. Para evitar problemas na plotagem de gráficos, usaremos apenas os dados de treino para as etapas de análise, já que apenas este contém a coluna 'is_promoted' preenchida; a limpeza, no entanto, ocorrerá no dataset 'uniao'. Ademais, estabeleceremos inicialmente duas funções que transformarão os dados em valores numéricos para auxiliar na montagem dos gráficos que se utilizam de dados categóricos. Observe abaixo:

```python
sns.set(style="darkgrid")    # Escurece o background dos gráficos
```

```python
def conta_ocorrencias(lista_de_possibilidades, feature):
    """ Função que conta o número de ocorrências de cada possibilidade em uma feature
    cujos valores são restritos a um conjunto finito de possibilidades.
    Entrada: uma lista com todas as possibilidades presentes em determinada feature e
    a indicação da feature que está sendo explorada.
    Saída: Uma lista de números em que cada valor corresponde ao número de ocorrências
    da possibilidade de mesmo lugar.                                               """
    lista_vazia = []
    for item in lista_de_possibilidades:    # Adiciona um 0 em 'lista_vazia' para cada palavra presente em 'lista_de_possibilidades', delimitando logo de início o tamanho da nova lista
        lista_vazia.append(0)
    for ocorrencia in feature:
        cont = 0
        while cont != len(lista_de_possibilidades):
            if ocorrencia == lista_de_possibilidades[cont]:    # Aumenta em 1 determinado item de 'lista_vazia' para cada ocorrência que ocupe a mesma posição em 'lista_de_possibilidades'
                lista_vazia[cont] = lista_vazia[cont] + 1
            cont = cont + 1
    return lista_vazia

def conta_duas_features(lista_de_possibilidades1, feature1, lista_de_possibilidades2, feature2):
    """ Função que conta o número de ocorrências de cada possibilidade em uma feature 1
    e divide tal número de acordo com o encaixe em uma feature 2, cujos valores estão 
    restritos a duas possibilidades.
    Entrada: uma lista com todas as possibilidades presentes em determinada feature 1,
    outra lista as duas possibilidades presentes em determinada feature 2 e a indicação
    das duas features que estão sendo exploradas.
    Saída: Uma lista de listas em que os valores correspondem ao número de ocorrências
    da possibilidade de mesmo lugar.                                               """
    lista_duas_features = []
    for item in lista_de_possibilidades1:    # Adiciona uma lista composta por dois zeros em 'lista_vazia' para cada palavra presente em 'lista_de_possibilidades1', delimitando logo de início o tamanho da nova lista
        lista_duas_features.append([0, 0])
    cont2 = 0
    for ocorrencia in feature1:
        cont1 = 0
        while cont1 != len(lista_de_possibilidades1):
            if ocorrencia == lista_de_possibilidades1[cont1]:    # Verifica qual a posição que uma determinada ocorrência ocupa em 'lista_de_possibilidades1' e depois vê a qual opção de 'lista_de_possibilidades2' a linha do dataset analisada corresponde
                if feature2[cont2] == lista_de_possibilidades2[0]:
                    lista_duas_features[cont1][0] = lista_duas_features[cont1][0] + 1
                else:
                    lista_duas_features[cont1][1] = lista_duas_features[cont1][1] + 1
            cont1 = cont1 + 1
        cont2 = cont2 + 1
    return(lista_duas_features)
```

### Lidando com os dados repetidos em 'gender' e os poucos dados faltantes de 'region' e 'department'
Pela análise dos valores apresentados, encontramos algumas categorias que são, no mínimo, estranhas. Devemos, portanto, unir algumas categorias repetidas (como no caso de 'gender', em que 'f' e 'fEMALE' possuem o mesmo significado) e verificar o número de ocorrências de categorias anormais para excluí-las do dataset, caso sua frequência seja irrisória.

```python
suspeitosDepartment = len(uniao[uniao['department'] == 'quem trouxe quenga'])
suspeitosRegion = len(uniao[uniao['region'] == 'casa da lu na suecia'])
suspeitosPercentual = (suspeitosDepartment + suspeitosRegion) / len(uniao) * 100
print('O número de ocorrências suspeitas é:', str(suspeitosDepartment + suspeitosRegion))   # Verifica o número de ocorrências de categorias suspeitas
print('A porcentagem de ocorrências suspeitas em relação ao dataset total é de:', str(suspeitosPercentual), '%')    # Verifica a porcentagem de categorias suspeitas em relação ao dataset
```

Considerando que, nesse caso, as 38 ocorrências de categorias suspeitas constituem menos de 0,05% do dataset, podemos optar por simplesmente deletar as linhas em que elas aparecem. Observe as linhas de código abaixo, em que são excluídos os valores suspeitos e unificadas as categorias repetidas:

```python
indexSuspeitos = uniao[uniao['region'] == 'casa da lu na suecia'].index
uniao.drop(indexSuspeitos, inplace=True)    # Exclui as 6 linhas com valores suspeitos na feature 'region'

indexSuspeitos = uniao[uniao['department'] == 'quem trouxe quenga'].index
uniao.drop(indexSuspeitos, inplace=True)    # Exclui as 32 linhas com valores suspeitos na feature 'department'

uniao['gender'].replace(('MALE', 'm', 'eu sou homem hetero homossexual!'), 'male', inplace=True)    # Na feature 'gender', substitui 'MALE', 'm' e 'eu sou homem hetero homossexual!' por 'male'
uniao['gender'].replace(('fEMALE', 'f'), 'female', inplace=True)    # Na feature 'gender', substitui 'fEMALE' e 'f' por 'female'
```

### Lidando com dados aberrantes em 'age'
Logo de início, alguns valores se destacam em nossa tabela, e muitos desses outliers provavelmente representam erros que devem ser substituídos ou excluídos de nosso dataset. 
Sabendo que o número máximo (123456) e mínimo (-32) presentes em 'age', por exemplo, estão claramente equivocados, vamos analisar a quantidade de valores aberrantes dessa categoria para decidir se iremos substituí-los ou apagá-los, caso sejam poucos.

```python
count = 0
for elemento in uniao['age']:    # Conta a quantidade de valores acima de 100 e abaixo de 18 presentes na coluna 'age'
    if elemento > 100 or elemento < 18:
        count = count + 1
print("O número de valores aberrantes presentes em 'age' é de:", count)
```

Considerando suas 109 aparições e sabendo que a média de idade seria de aproximadamente 35 anos caso esses valores discrepantes não estivessem presentes, vamos substituir tais números pela média de idades para não perder mais dados do dataset:

```python
uniao.isnull().sum()
```

```python
for elemento in uniao['age']:    # Substitui as idades acima de 100 e abaixo de 18 pela média 35
    if elemento > 100 or elemento < 0:
        uniao['age'].replace(elemento, 35, inplace=True)
```

## Modelo de regressão logística
Iremos, inicicialmente, importar o modelo pelo seguinte comando:

```python
from sklearn.linear_model import LogisticRegression    # Modelo que usaremos
```

Em seguida, separaremos os dados de treino em X (por convenção, o 'X' é escrito em caixa alta) e y, sendo que o primeiro representa as colunas usadas para fazer as predições e o segundo a coluna target, ou aquela que queremos predizer.

```python
X_treino = novotreino.drop(columns = ['employee_id', 'department', 'region', 'education', 'gender', 'recruitment_channel', 'no_of_trainings', 'length_of_service', 'awards_won?', 'is_promoted'])    # Cria uma variável apenas com os valores mais importantes para prever o target (isto é, 'age', 'previous_year_rating' e 'avg_training_score')
y_treino = novotreino['is_promoted']    # Cria uma variável apenas com os valores do target

X_teste = novoteste.drop(columns = ['employee_id', 'department', 'region', 'education', 'gender', 'recruitment_channel', 'no_of_trainings', 'length_of_service', 'awards_won?', 'is_promoted'])    # Cria uma variável apenas com os valores mais importantes para prever o target (isto é, 'age', 'previous_year_rating' e 'avg_training_score')
```

Agora, vamos aplicar esse modelo para os nossos dados de treino e de teste, predizendo o valor de nosso target:

```python
logreg = LogisticRegression()    # Cria um modelo de regressão logística

logreg.fit(X_treino, y_treino)
y_treino_predict = logreg.predict(X_treino)    # Prediz os dados de treino
y_teste_predict = logreg.predict(X_teste)    # Prediz os dados de teste
```

Por fim, podemos ainda avaliar o modelo por diferentes métricas. Lembrando que limitaremos a avaliação apenas para os dados de treino, visto que, nos dados de teste, a falta de um valor atribuído a nossa target impossibilita a verificação da qualidade de predição. Veja abaixo:

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
acuracia = accuracy_score(y_treino, y_treino_predict)    # Avalia a acurácia nos dados de treino
print('Acurácio no treino:', acuracia)

precisao = precision_score(y_treino, y_treino_predict, average='weighted')    # Avalia a precisão nos dados de treino
print('Precisão no treino:', precisao) 

revocacao = recall_score(y_treino, y_treino_predict, average='weighted')    # Avalia a revocação nos dados de treino
print('Revocação no treino:', revocacao)
```

Pela observação das métricas acima vemos que, a acurácia, apesar de ser uma métrica simples e intuitiva, não seria a melhor avaliadora em nosso caso mesmo apresentando um alto valor, visto que os dados estão desbalanceados (ou seja, o número de funcionários não promovidos é muito maior que aqueles que foram promovidos). Assim, são introduzidas as métricas precisão (concede maior peso para falsos positivos) e revocação (concede maior peso para falsos negativos), capazes de captar mais precisamente a real qualidade de nosso modelo. Visto que ambas as grandezas encontram-se elevadas, concluímos que nosso modelo é capaz de prever o target de maneira adequada e balanceada.
